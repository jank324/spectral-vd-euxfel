```python
import pickle

import matplotlib.pyplot as plt
import scienceplots
```

```python
plt.style.use(["science", "ieee"])
```

```python
data_for_sergey = pickle.load(open("data_for_sergey.pkl", "rb"))
```

```python
num_examples = 4
scale = 1.25

print(f"{plt.rcParams['xtick.labelsize'] = }")
print(f"{plt.rcParams['font.size'] = }")
# print(f"{plt.rcParams = }")

plt.figure(figsize=(6.69423 * scale, 6.69423 * scale * 0.5 * 0.38))

for i in range(num_examples):
    ground_truth = data_for_sergey[f"ground_truth_{i}"]
    neural_network = data_for_sergey[f"neural_network_{i}"]

    plt.subplot(1, num_examples, i + 1)
    plt.plot(ground_truth["s_m"] * 1e3, ground_truth["I_A"], label="Ground truth")
    plt.plot(neural_network["s_m"] * 1e3, neural_network["I_A"], label="Reconstruction")
    # plt.plot(recon_time, recon_current)
    plt.xlabel(r"$s$ (mm)")
    if i == 0:
        plt.ylabel(r"$I$ (A)")
    if i == num_examples - 1:
        plt.legend()


plt.tight_layout()
# plt.subplots_adjust(wspace=0.4, hspace=0.3)
# plt.savefig("figures/THPL019_f2.pdf")
plt.show()
```
