```python
from pathlib import Path

import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data_dir = Path("data/zihan")
```

```python
df_full = pd.read_pickle(data_dir / "data_20220905_full.pkl")
df_full
```

```python
df_train, df_val_test = train_test_split(df_full, test_size=0.2)
df_val, df_test = train_test_split(df_val_test, test_size=0.5)
```

```python
len(df_train), len(df_val), len(df_test)
```

```python
df_train.to_pickle(data_dir / "data_20220905_train.pkl")
df_val.to_pickle(data_dir / "data_20220905_validation.pkl")
df_test.to_pickle(data_dir / "data_20220905_test.pkl")
```
