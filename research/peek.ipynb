```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open("tds_feb_phase1.pkl", "rb") as file:
    phase1 = pickle.load(file)
```

```python
get_peak_value = lambda x: x[1].max()
phase1.apply(get_peak_value).plot()
plt.xlabel("L1 Chirp Phase")
plt.ylabel("Peak Current (A)")
plt.show()
```
